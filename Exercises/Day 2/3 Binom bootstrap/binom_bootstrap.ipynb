### Exercise 3

```python
N = np.random.binomial(100,0.19,1000)

estimates = []
for i in N:
    estimates.append(i/100)

estimates,np.mean(estimates)
```

### Exercise 4

```python
plt.hist(estimates,20)
```

### Exercise 5

```python
from scipy.stats import bootstrap
estimate = (estimates,)
bootstrap_ci = bootstrap(estimate,np.mean,confidence_level=0.95,random_state=1,method="percentile")
bootstrap_ci.confidence_interval
```
